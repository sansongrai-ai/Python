```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
```

```python
plt.title("Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")

# Step 4: Show plot
plt.show()
```
