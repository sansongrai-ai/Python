```python
import pandas as pd

# Load dataset
df = pd.read_csv(r"C:\Users\kesha\Downloads\AB_NYC_2019.csv")

# Display basic information
df.info()
```

```python
sample_df = df.sample(n=1000, random_state=42)

print(sample_df.shape)
```

```python
sample_df["reviews_per_month"] = sample_df["reviews_per_month"].fillna(0)
sample_df["host_name"] = sample_df["host_name"].fillna("Unknown")
sample_df["name"] = sample_df["name"].fillna("No Name")
print(sample_df["reviews_per_month"])
print(sample_df["host_name"])
print(sample_df["name"])
```

```python
sample_df["last_review"] = sample_df["last_review"].fillna("Not Reviewed")
print(sample_df["last_review"])
```

```python
sample_df.drop_duplicates(inplace=True)
sample_df.duplicated().sum()
```

```python
Q1 = sample_df["minimum_nights"].quantile(0.25)
Q3 = sample_df["minimum_nights"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

night_outliers = sample_df[
    (sample_df["minimum_nights"] < lower) |
    (sample_df["minimum_nights"] > upper)
]

print(len(night_outliers))
```

```python
Q1 = sample_df["price"].quantile(0.25)
Q3 = sample_df["price"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

price_outliers = sample_df[
    (sample_df["price"] < lower) |
    (sample_df["price"] > upper)
]

print(len(price_outliers))
```

```python
sample_df.drop("id", axis=1, inplace=True)
sample_df.drop("host_id", axis=1, inplace=True)
sample_df.drop("host_name", axis=1, inplace=True)
```
