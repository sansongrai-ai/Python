```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\kesha\Downloads\customer_purchase_behavior.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df["SpendingScore"], bins=10, kde=True)
plt.title("Histogram of Spending Score")
plt.xlabel("Spending Score")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df["AnnualIncome"], bins=10, kde=True)
plt.title("Histogram of Annual Income")
plt.xlabel("Annual Income")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="CustomerType", y="SpendingScore", data=df)
plt.title("Spending Score by Customer Type")
plt.xlabel("Customer Type")
plt.ylabel("Spending Score")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(
    x="AnnualIncome",
    y="SpendingScore",
    hue="CustomerType",
    data=df
)
plt.title("Annual Income vs Spending Score")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(
    x="Age",
    y="SpendingScore",
    hue="CustomerType",
    data=df
)
plt.title("Age vs Spending Score")
plt.xlabel("Age")
plt.ylabel("Spending Score")
plt.show()
```
